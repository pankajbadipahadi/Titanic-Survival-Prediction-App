# **Task 2 : Data Preprocessing**

**Handle missing values (e.g., imputation).**

```python
display(df_train.isnull().sum())
display(df_test.isnull().sum())
```

```python
# Impute missing Age values with the median
df_train['Age'].fillna(df_train['Age'].median(), inplace=True)

# Impute missing Embarked values with the mode
df_train['Embarked'].fillna(df_train['Embarked'].mode()[0], inplace=True)

# Display handled missing values
display(df_train.isnull().sum())
```

```python
# Impute missing Age values with the median
df_test['Age'].fillna(df_test['Age'].median(), inplace=True)

# Impute missing Embarked values with the mode
df_test['Fare'].fillna(df_test['Fare'].median(), inplace=True)

# Display handled missing values
display(df_test.isnull().sum())
```

**Encode categorical variables.**

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df_train['Sex']=le.fit_transform(df_train['Sex'])
df_train['Embarked']=le.fit_transform(df_train['Embarked'])
df_train
```

```python
features=['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']
x=df_train[features]
y=df_train['Survived']
```

```python
print("Features used:", features)
```

# **Task : 3. Model Building**

```python
from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
# Scaling the features
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_val_scaled=scaler.transform(x_val)
```

**Build a logistic regression model using appropriate libraries (e.g., scikit-learn).**

```python
print("\n Training logistic regression model...")
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=42, max_iter=1000)
model.fit(x_train_scaled, y_train)
```

**Train the model using the training data.**

```python
model.score(x_train_scaled,y_train)*100
```

```python
# Make predictions
y_pred = model.predict(x_val_scaled)
display(y_pred)
```

# **Task : 4. Model Evaluation**

**Evaluate the performance of the model on the testing data using accuracy, precision, recall, F1-score, and ROC-AUC score.**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,classification_report, roc_auc_score

# Make predictions on the validation set
y_pred_val = model.predict(x_val_scaled)
y_pred_proba_val = model.predict_proba(x_val_scaled)[:, 1]

# Evaluate the model
accuracy = accuracy_score(y_val, y_pred_val)
precision = precision_score(y_val, y_pred_val)
recall = recall_score(y_val, y_pred_val)
f1 = f1_score(y_val, y_pred_val)
auc_score = roc_auc_score(y_val, y_pred_proba_val)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"ROC-AUC Score: {auc_score:.4f}")

print("\nClassification Report:")
print(classification_report(y_val, y_pred_val))
```

**Visualize the ROC curve.**

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Calculate the false positive rate, true positive rate, and thresholds
fpr, tpr, thresholds = roc_curve(y_val, y_pred_proba_val)

# Calculate the ROC AUC score
roc_auc = roc_auc_score(y_val, model.predict_proba(x_val_scaled)[:, 1])

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.grid(True)
plt.show()
```

# **Task : 5. Interpretation**

**Interpret the coefficients of the logistic regression model.**

```python
print("\n Model Interpretation:")
coefficients = model.coef_[0]

# Get the feature names used for training
features = x_train.columns

coef_df = pd.DataFrame({
    'Feature': features,
    'Coefficient': coefficients,
    'Odds_Ratio': np.exp(coefficients),
    'Abs_Effect': np.abs(coefficients)
}).sort_values('Abs_Effect', ascending=False)

print("Feature coefficients and odds ratios:")
display(coef_df)

# Feature importance plot
plt.figure(figsize=(10, 6))
colors = ['red' if x < 0 else 'green' for x in coef_df['Coefficient']]
plt.barh(coef_df['Feature'], coef_df['Abs_Effect'], color=colors)
plt.xlabel('Absolute Coefficient Value')
plt.title('Feature Importance')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

print("\nKey Insights:")
print("- Positive coefficients increase survival probability")
print("- Negative coefficients decrease survival probability")
print("- The magnitude of the absolute coefficient indicates the feature's importance")
```

**Discuss the significance of features in predicting the target variable (survival probability in this case).**

Ans:- Based on the logistic regression model and the feature importance plot:

*   **Sex (Gender)** has the largest absolute coefficient, indicating it is the most significant predictor of survival. The negative coefficient and odds ratio less than 1 for 'Sex' (assuming 'male' was encoded as 1 and 'female' as 0) suggest that being male significantly decreases the probability of survival compared to being female.
*   **Pclass (Passenger Class)** is the second most important feature. The negative coefficient and odds ratio less than 1 suggest that higher passenger classes (e.g., Pclass 3) had a lower probability of survival compared to lower classes (e.g., Pclass 1).
*   **Age** is also a significant feature, with a negative coefficient and odds ratio less than 1, indicating that older individuals had a lower probability of survival.
*   **SibSp (Number of Siblings/Spouses Aboard)** has a negative coefficient, suggesting that having more siblings or spouses decreased the probability of survival.
*   **Fare** has a positive coefficient and odds ratio greater than 1, indicating that a higher fare is associated with a higher probability of survival.
*   **Embarked (Port of Embarkation)** has a negative coefficient, suggesting that the port of embarkation had some influence on survival probability.
*   **Parch (Number of Parents/Children Aboard)** has a coefficient close to zero, suggesting it had minimal impact on survival probability in this model.

In summary, gender, passenger class, and age appear to be the most influential factors in predicting survival on the Titanic according to this logistic regression model.
